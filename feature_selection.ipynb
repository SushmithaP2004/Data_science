```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import ExtraTreesClassifier
df=pd.read_csv('googleplaystore (2).csv')
df
df_numeric=df.select_dtypes(include=['number','float64'])
```

```python
df.isnull().sum()
```

```python
df[['Rating','Content Rating','Current Ver','Android Ver']]=df[['Rating','Content Rating','Current Ver','Android Ver']].ffill()
```

```python
df.isnull().sum()
```

```python
y=df['Genres']
X=df.drop('Genres',axis=1)
X.head()
```

```python
extra_tree_forest=ExtraTreesClassifier(n_estimators=5,criterion='entropy',max_features=2)
extra_tree_forest.fit(df_numeric,y)


```

```python
#feature importance
feature_importance=extra_tree_forest.feature_importances_
```

```python
plt.bar(df_numeric.columns,feature_importance)
plt.xlabel('Feature labels')
plt.ylabel('Feature importance')
plt.title('comparison of different feature importances')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(df)
scaled_data=scalar.transform(df)
```

```python
#importing pca
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
x_pca.shape
```

```python

plt.figure(figsize =(8, 6))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c = cancer['target'], cmap ='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

```python
df_comp=pd.DataFrame(pca.components_,columns=cancer['feature_names'])
plt.figure(figsize=(6,8))
sns.heatmap(df_comp)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
#import standardscalar module
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(df)
scaled_data=scalar.transform(df)
```

```python
#importing pca
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(scaled_data)
x_pca=pca.transform(scaled_data)
x_pca.shape
```

```python
plt.figure(figsize=(6,8))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('first principal component')
plt.ylabel('second principal component')
```

```python
pca.components_
```

```python
df_comp=pd.DataFrame(pca.components_,columns=cancer['feature_names'])
plt.figure(figsize=(14,6))
sns.heatmap(df_comp)
```
