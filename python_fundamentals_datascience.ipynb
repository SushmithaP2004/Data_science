```python
print("welcom to student management system")
student={}
while True:
  print("menu")
  print("1.add student")
  print("2.view all students")
  print("3.update marks")
  print("4.exit")

  choice=int(input("enter your choice"))
  if(choice==1):
    name=input("enter name")
    marks=int(input("enter marks:"))
    student[name]=marks
    print(f"{name}added with {marks} marks")
  elif choice==2:
    if not student:
      print("no student to be added")
    else:
      print("student list")
  elif choice==3:
    name=input("enter student name to update")
    if name in student:
      new_marks=int(input("enter new marks"))
      student[name]=new_marks
      print(f"{name} s marks updated to{new_marks}")
    else:
      print("student not found")
  elif choice==4:
    print("exiting")
    break
  else:
    print("invalid choice please try again")
```

```python
import os

def file_create(file):
    try:
        with open(file, "x") as s:
            pass
        print("File created successfully")
    except FileExistsError:
        print("File already exists")

def file_write(file, content):
    try:
        with open(file, "w") as s:
            s.write(content)
        print("Successfully written")
    except Exception as e:
        print("Error:", e)

def file_append(file, content):
    try:
        with open(file, "a") as s:
            s.write(content)
        print("Successfully appended")
    except Exception as e:
        print("Error:", e)

def file_read(file):
    try:
        with open(file, "r") as s:
            print(s.read())
        print("Read successful")
    except Exception as e:
        print("Error in reading content:", e)

def file_delete(file):
    try:
        os.remove(file)
        print("File removed successfully")
    except FileNotFoundError:
        print("File doesn’t exist")

# Menu loop
while True:
    print("\nMenu")
    print("1. Create File")
    print("2. Write File")
    print("3. Append File")
    print("4. Read File")
    print("5. Delete File")
    print("6. Exit")

    try:
        ch = int(input("Enter your choice: "))
    except ValueError:
        print("Please enter a number between 1-6")
        continue

    if ch == 1:
        a = input("Enter the file name: ")
        file_create(a)
    elif ch == 2:
        a = input("Enter filename: ")
        y = input("Enter the content: ")
        file_write(a, y)
    elif ch == 3:
        a = input("Enter filename: ")
        y = input("Enter the content: ")
        file_append(a, y)
    elif ch == 4:
        x = input("Enter filename: ")
        file_read(x)
    elif ch == 5:
        x = input("Enter filename: ")
        file_delete(x)
    elif ch == 6:
        print("Exit")
        break
    else:
        print("Please enter a choice between 1-6")
```
